# WebScrapt all The Topics in Forum

```python
import urllib
from bs4 import BeautifulSoup

url="Day 3, Section 3 (I-L) Discussion- Nativity Scene.htm"
page = urllib.urlopen(url).read()
soup = BeautifulSoup(page)
forum=soup.find_all("div",class_="discussion-body layout layout-1t2t")[0]
forum_nav_thread=forum.find_all("ul",class_="forum-nav-thread-list")[0]
forum_nav=forum_nav_thread.find_all("li",class_="forum-nav-thread")
topics=[]
cl="forum-nav-thread-title"
for link in soup.find_all("span",class_=cl):
    topics.append(link.text)
    
all_info={}
for i in xrange(len(forum_nav)):
    info={}
    section=forum_nav[i]
    idx=section["data-id"]
    
    info["href"]=section.find_all("a",class_="forum-nav-thread-link")[0]["href"].encode("utf-8")
    info["title"]=section.find_all("span",class_="forum-nav-thread-title")[0].text
    info["count"]=int(section.find_all("span",class_="forum-nav-thread-comments-count")[0].get_text().strip().encode("utf-8"))
    all_info[idx]=info    
```

```python
import pickle
pickle.dump( all_info, open( "topics.p", "wb" ) )
```

### Choose One Topic and scrapt the Content of the Topic and all of the comments

```python
url="Day 3, Section 3 (I-L) Discussion- Nativity Scene.htm"

import urllib

page = urllib.urlopen(url).read()
soup = BeautifulSoup(page)

main=soup.find_all("main",class_="discussion-column layout-col layout-col-b")[0]
thread_wrapper=main.find_all("div",class_='thread-wrapper')[0]

main_wrapper=thread_wrapper.find_all("div",class_='thread-main-wrapper')[0]
discussion_post=main_wrapper.find_all("div",class_='discussion-post')[0]
title=discussion_post.find_all("div",class_='post-header-content')[0].find_all("h1",class_='post-title')[0].text.encode("utf-8")
post_body=discussion_post.find_all("div",class_='post-body')[0]
post_content=""
for link in post_body.find_all("p"):
    post_content+=link.text.encode("utf-8")   
    
extended_content=thread_wrapper.find_all("div",class_='post-extended-content thread-responses-wrapper')[0]    
responses=thread_wrapper.find_all("li",class_="forum-response")  


response_text={}
for i in xrange(len(responses)):
    response=responses[i].find_all("div",class_="response-body")
    contents=[]
    for r in response:
        content=""
        for p in r.find_all("p"):
            txt=p.text
            content+=txt
        contents.append(content)
    response_text[i]=  contents  
    
info={}
info["topics_content"]=post_content
info["comments"]=response_text    

```
